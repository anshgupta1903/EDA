```python
# 1 MB = 1024 Kb

def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M", ""))
        res = res * 1024
        return res
    elif str(item)[-1] == "k":
        res = float(str(item).replace("k", ""))
        return res
    else:
        return str(np.NaN)
```

```python
# example
size_process("19M")
```

```python
df_copy["Size"] = df_copy["Size"].apply(size_process)
```

```python
def android_var(item):
    item = str(item) # conver object to str
    if '-' in item:
        res = item.replace("-", "")
        x = res.split()  # Convert to list of integers
        res = max(x)  # Get max value
        return res
        
    elif item.count('.') > 1:
        res = item[:-2]  # Remove the last two characters
        return res
    else:
        res = item.replace('W', '')  # Remove 'W'
        return res
```

```python
android_var("2.3.3")
```

```python
df_copy["Android Ver"] =df_copy["Android Ver"].apply(android_var)
```
