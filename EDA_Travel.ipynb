# train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('ProdTaken', axis = 1)
y = df["ProdTaken"]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
# data is either numerical or categorical, numerical can be either discrete and continuous
# discrete data though its numerical but should be treated as categorical to do analysis
# but for model building all the categorical data should be converted to numbers
```

```python
cat_features = X.select_dtypes(include = 'object').columns
num_features = X.select_dtypes(exclude = 'object').columns
```

```python
num_features
```

```python
# Scaling >> numerical feature
# Encoding >> Categorical Data
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler #both are data transformations
from sklearn.compose import ColumnTransformer

# ColumnTransformer applies transformations to columns of an array or pandas dataframe
```

```python
numeric_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop = 'first')
preprocessor = ColumnTransformer(
    [
        ('OneHotEncoder', oh_transformer,cat_features),
        ('StandardScaler', numeric_transformer, num_features)
    ]
)
```

```python
preprocessor
```

```python
X_train = preprocessor.fit_transform(X_train) # execute only once because it 
# works on only dataframes and once executed it converts the dataframe to numpy array
```
