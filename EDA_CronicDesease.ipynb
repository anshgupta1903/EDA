```python
df['diabetes_mellitus'].unique()
```

```python
df['coronary_artery_disease'].replace(to_replace = '\tno', value = 'no', inplace = True)
```

```python
df['class'] = df['class'].map({'ckd' : 1, 'notckd' : 0, 'ckd\t' : 1})
# you could have also first replaced and then map with 1 and 0
```

# Multivariate Analysis

```python
col = ['age', 'blood_pressure', 'blood_glucose_random', 'serum_creatinine', 'class']
```

```python
df[col]
```

```python
graph = sns.PairGrid(df[col],hue = 'class', palette = 'coolwarm')
graph.map_upper(sns.scatterplot) # above the diagnol
graph.map_lower(sns.kdeplot, cmap = 'Blues_d') # below the diagnol
graph.map_diag(sns.histplot)
graph.add_legend()
plt.title("Pairgrid for selected columns")
plt.show()
```
